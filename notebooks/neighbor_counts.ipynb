```python
import numpy as np
import matplotlib.pyplot as plt

import os
import sys
sys.path.append(os.path.join(os.path.abspath(''), '..'))
from utils.dla import get_neighbors, new_position, step
```

```python
aggragations = 1000
size = 150

grid = np.zeros((size,size))
grid[size//2, size//2] = 1

current = new_position(grid)
neighbors = get_neighbors(grid)

grid_list = [grid]
i = 0
while i < aggragations:
    current, stopped = step(grid, current, neighbors)
    i += stopped
    if stopped:
        grid[*current] = 1
        grid_list.append(grid.copy())
        current = new_position(grid)
        neighbors = get_neighbors(grid)
        print(f'\r{i}/{aggragations}', end='')

plt.imshow(grid);
```
